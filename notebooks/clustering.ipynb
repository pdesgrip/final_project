```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import joblib
```

```python
# load dataset
taxi_data = pd.read_csv('../data/yellow/taxi_y_cleaned.csv')

# show first few lines
taxi_data.head()
```

```python
# check for missing values
missing_values = taxi_data.isnull().sum()

# get basic statistical details
stats = taxi_data.describe()

missing_values, stats
```

```python
# check for missing values
missing_values = taxi_data.isnull().sum()

# get basic statistical details
stats = taxi_data.describe()

missing_values, stats
```

```python
# define a function to remove outliers based on IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# remove negative values for fare_amount and tolls_amount by removing such records
taxi_data_cleaned = taxi_data[(taxi_data['fare_amount'] >= 0) & (taxi_data['tolls_amount'] >= 0)]

# remove rate_code outlier
taxi_data_cleaned = taxi_data_cleaned[taxi_data_cleaned['ratecodeid'].isin(range(1, 7))]

# checkc extreme values in trip_duration_seconds by removing outliers using the IQR 
taxi_data_cleaned = remove_outliers(taxi_data_cleaned, 'trip_duration_seconds')

# summary of data after cleaning
summary_after_cleaning = {
    "Initial_Count": len(taxi_data),
    "Count_After_Negative_Fare_Removal": len(taxi_data_cleaned),
    "Count_After_RateCodeID_Fix": len(taxi_data_cleaned),
    "Count_After_Duration_Outlier_Removal": len(taxi_data_cleaned)}

summary_after_cleaning
```

```python
# separate the features and potential labels
features = taxi_data_cleaned.drop(['payment_type'], axis=1)  # payment_type could be a proxy for disputes
labels = taxi_data_cleaned['payment_type']

# identify numerical and categorical columns
numerical_cols = features.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = features.select_dtypes(include=['object', 'bool', 'category']).columns

# transfirm columns
preprocessor = ColumnTransformer(transformers=[('num', StandardScaler(), numerical_cols), 
                                               ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)])

# make pipelne 
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# fit and xform
features_prepared = pipeline.fit_transform(features)

# shape
features_prepared.shape
```

We will attempt to reduce features. Attemped with PCA but it doesn't seem compatible with OneHotEncoded feats. Looked up SVD.

```python
# initialise with 50 components
svd = TruncatedSVD(n_components=50)

# fit and xform using SVD
features_svd = svd.fit_transform(features_prepared)

# explain variance ratio
explained_variance = svd.explained_variance_ratio_.sum()

# manually set no. components
n_components = 50

n_components, explained_variance, features_svd.shape
```

We reduce our feature set whilst retaining over 80% of the predictive power

```python
# calculates the sum of squared distances of samples to their closest cluster center
def calculate_wcss(data):
    wcss = []
    for n in range(2, 11):
        kmeans = KMeans(n_clusters=n, random_state=33)
        kmeans.fit(data)
        wcss.append(kmeans.inertia_)
    return wcss

# calculate the cluster(within) sum of squares for different numbers of clusters
wcss = calculate_wcss(features_svd)

# plot elbow
plt.figure(figsize=(10, 8))
plt.plot(range(2, 11), wcss, marker='o', linestyle='--')
plt.title('K-means Clustering: The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # within cluster sum of squares
plt.show()
```

It looks to me like 6 is the optimal number of clusters. Rate of decline of slope (elbow) is there.

```python
# execute KMeanswith 6 clusters
kmeans = KMeans(n_clusters=6, random_state=33)
kmeans.fit(features_svd)

# get labels
cluster_labels = kmeans.labels_

# add to df
taxi_data_cleaned['cluster'] = cluster_labels

# take a peek 
taxi_data_cleaned.head()
```

```python
# add labels to df
taxi_data_cleaned['cluster'] = cluster_labels

# analyze the clusters
cluster_summary = taxi_data_cleaned.groupby('cluster').agg({
    'fare_amount': ['mean', 'median', 'std'],
    'tip_amount': ['mean', 'median', 'std'],
    'total_amount': ['mean', 'median', 'std'],
    'trip_distance': ['mean', 'median', 'std'],
    'trip_duration_seconds': ['mean', 'median', 'std'],
    'payment_type': lambda x: x.value_counts().index[0] #most common payemnt_type
}).reset_index()

# flatten index
cluster_summary.columns = ['_'.join(col).strip() if col[1] else col[0] for col in cluster_summary.columns.values]

cluster_summary
```

```python
# convert 'trip_duration_seconds_mean' to minutes for better comparison
cluster_summary['trip_duration_seconds_mean'] /= 60
cluster_summary.rename(columns={'trip_duration_seconds_mean': 'trip_duration_minutes_mean'}, inplace=True)

# melf the modified cluster summary for visualization
cluster_summary_melted = pd.melt(cluster_summary, 
                                 id_vars='cluster', 
                                 value_vars=['fare_amount_mean', 'tip_amount_mean', 'total_amount_mean', 'trip_distance_mean', 'trip_duration_minutes_mean'],
                                 var_name='Metric', 
                                 value_name='Value')

# create the updated cluster comparison plot
plt.figure(figsize=(14, 7))
sns.barplot(x='Metric', y='Value', hue='cluster', data=cluster_summary_melted)
plt.title('Comparison of Clusters Across Different Metrics (Duration in Minutes)')
plt.ylabel('Mean Value')
plt.legend(title='Cluster')
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

```python

labels = kmeans.labels_

# silhouette Score
silhouette_avg = silhouette_score(features_svd, labels)
print(f"Silhouette Score: {silhouette_avg}")

# Davies-Bouldin Index
db_score = davies_bouldin_score(features_svd, labels)
print(f"Davies-Bouldin Index: {db_score}")

# Calinski-Harabasz Index
ch_score = calinski_harabasz_score(features_svd, labels)
print(f"Calinski-Harabasz Index: {ch_score}")
```
