```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the cleaned CSV
taxi_data = pd.read_csv('../data/yellow/taxi_y_cleaned.csv')
```

```python
# convert pickup datetime to datetime object and extract the hour
taxi_data['tpep_pickup_datetime'] = pd.to_datetime(taxi_data['tpep_pickup_datetime'])
taxi_data['hour_of_day'] = taxi_data['tpep_pickup_datetime'].dt.hour
taxi_data['day_of_week'] = taxi_data['tpep_pickup_datetime'].dt.dayofweek
taxi_data['week_of_year'] = taxi_data['tpep_pickup_datetime'].dt.isocalendar().week
taxi_data['is_weekend'] = taxi_data['day_of_week'].isin([5, 6]).astype(int)
taxi_data['distance'] = taxi_data['trip_distance']
taxi_data['number_of_passengers'] = taxi_data['passenger_count']  
```

```python
# aggregate data to count the number of rides per hour and day of the week
hourly_weekly_ride_counts = taxi_data.groupby(['day_of_week', 'hour_of_day']).size().reset_index(name='ride_count')
```

```python
# distribution of ride count
sns.histplot(hourly_weekly_ride_counts['ride_count'], kde=True)
plt.title('Distribution of Ride Counts')
plt.xlabel('Ride Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# aggregate mean distance and mean passenger count
aggregated_data = taxi_data.groupby(['day_of_week', 'hour_of_day']).agg({
    'trip_distance': 'mean',   # avg distance
    'passenger_count': 'mean',  # avg passengers
}).reset_index()

# join with ride count
hourly_weekly_ride_counts = hourly_weekly_ride_counts.merge(aggregated_data, on=['day_of_week', 'hour_of_day'])

# rename the columns for clarity
hourly_weekly_ride_counts.rename(columns={
    'trip_distance': 'average_distance',
    'passenger_count': 'average_passenger_count'
}, inplace=True)

# Now, prepare your features (X) and target (y)
X = hourly_weekly_ride_counts.drop('ride_count', axis=1)
y = hourly_weekly_ride_counts['ride_count']

```

```python
# splits the data into train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=33)
```

```python
# random forest
random_forest = RandomForestRegressor(random_state=33)

# Define the parameter grid for randomized search
param_distributions = {
    'n_estimators': [100, 200, 500],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# start search model cross validation
random_search = RandomizedSearchCV(estimator=random_forest, param_distributions=param_distributions, n_iter=108, cv=10, verbose=2, random_state=42, n_jobs=-1)

# fir data
random_search.fit(X_train, y_train)

# display best params
display(f"Best parameters: {random_search.best_params_}")
best_rf_model = random_search.best_estimator_


```

```python
# evaluate the best model on the test set
from sklearn.metrics import mean_squared_error

y_pred = best_rf_model.predict(X_test)
rmse = mean_squared_error(y_test, y_pred, squared=False)
display(f"Test set RMSE: {rmse}")
```

```python
# save the best model from randomized search
joblib.dump(best_rf_model, '../models/random_forest_model_improved.pkl')
```

```python
# feature importance
feature_importances = pd.Series(best_rf_model.feature_importances_, index=X.columns)
sns.barplot(x=feature_importances, y=feature_importances.index)
plt.title('Feature Importances')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.show()
```
